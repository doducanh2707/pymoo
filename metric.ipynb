```python
def read_data(file_name):
    inds = []
    with open(file_name) as f:
        data = f.readlines()
    for line in data:
        f1,f2 = line.split(' ')
        inds.append((f1,f2))
    return inds
```

```python
def dominate(a,b):
    for i in range(2):
        if a[i] > b[i]:
            return False
    return True 
```

```python
def c_metric(filename_a,filename_b):
    algo_a = np.loadtxt(filename_a)
    algo_b = np.loadtxt(filename_b)
    c = 0
    for a in algo_a:
        for b in algo_b:
            if dominate(b,a):
                c+=1
                break
                
    return c / len(algo_a)
```

```python
from pymoo.indicators.hv import Hypervolume
def hv_metric(filename_a,filename_b):
    algo_a = np.loadtxt(filename_a)
    algo_b = np.loadtxt(filename_b)    
    approx_ideal = algo_a.min(axis=0)
    approx_nadir = algo_a.max(axis=0)
    metric = Hypervolume(ref_point= np.array([1, 1]), 
                        norm_ref_point=False,
                        zero_to_one=False, 
                        ideal=approx_ideal,
                        nadir=approx_nadir)
    hv_a = metric.do(algo_a)
    approx_ideal = algo_b.min(axis=0)
    approx_nadir = algo_b.max(axis=0)
    metric = Hypervolume(ref_point= np.array([1, 1]), 
                        norm_ref_point=False,
                        zero_to_one=False, 
                        ideal=approx_ideal,
                        nadir=approx_nadir)
    hv_b = metric.do(algo_b)
    return hv_a - hv_b
```
