```python
import os 
import glob 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
def dominate(a,b):
    cnt = 0 
    for i in range(2):
        if a[i] > b[i]:
            return False
        if a[i] == b[i]:
            cnt+=1
    if cnt == 2:
        return False
    return True 
```

```python
def c_metric(filename_a,filename_b):
    algo_a = np.unique(np.loadtxt(filename_a),axis= 0)
    algo_b = np.unique(np.loadtxt(filename_b),axis= 0)
    c = 0
    for x in algo_b:
        for y in algo_a:
            if dominate(y,x):
                c+=1
            break
    
    return c / len(algo_b)
```

```python
from pymoo.indicators.hv import Hypervolume
def hv_metric(filename_a):
    algo_a = np.unique(np.loadtxt(filename_a),axis = 0)
    metric = Hypervolume(ref_point= np.array([1.1, 1.1]))
    hv_a = metric.do(algo_a)
    return hv_a 
```

```python
def nds(filename):
    return len(np.unique(np.loadtxt(filename),axis = 0))
```

```python
for instance in os.listdir('result/nsga2'):
    a = f'result/nsga2_custom/{instance}'
    b = f'result/moead/{instance}'
    print(round(c_metric(a,b) - c_metric(b,a),2))
```
